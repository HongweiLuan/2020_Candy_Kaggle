```python
cols = list(df3.columns)
cols.remove('team')
cols.remove('bandit')
cols.remove('machine_id')
TRAIN_FEATS = cols
TARGET_COL = 'bandit'
```

```python
df3[TARGET_COL] = df3[TARGET_COL].apply(lambda x : 100 if x == -1 else x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df3[TRAIN_FEATS], df3[TARGET_COL], test_size=0.2, random_state=42)
```

```python

import lightgbm as lgb
# dtrain = lgb.Dataset(train_df[TRAIN_FEATS], label=train_df[TARGET_COL])
dtrain = lgb.Dataset(train_df[TRAIN_FEATS], label=train_df[TARGET_COL])
dtest = lgb.Dataset(test_df[TRAIN_FEATS], label=test_df[TARGET_COL])
param = {'objective': 'regression', 'num_leaves': 31, 'learning_rate': 0.09}
param['metric'] = 'rmse'

num_round = 1200

bst = lgb.train(param, dtrain, num_round, valid_sets=[dtrain, dtest], early_stopping_rounds=5, verbose_eval=20)
```

```python
lgb.plot_importance(bst, importance_type='gain', max_num_features=30)
```
