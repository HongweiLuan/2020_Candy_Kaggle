## Get Topteam(nagiss) Episode

```python
def saveEpisode(sub_id, epid, rj):
    # request
    re = requests.post(get_url, json = {"EpisodeId": int(epid)})
        
    # save replay
    with open('{}_{}.json'.format(sub_id, epid), 'w') as f:
        f.write(re.json()['result']['replay'])

    # save episode info
    with open('{}_{}_info.json'.format(sub_id, epid), 'w') as f:
        json.dump([r for r in rj['result']['episodes'] if r['id']==epid][0], f)
```

```python
sub = 18669376 # nagiss's public LB submission ID

start_time = datetime.datetime.now()
r = BUFFER;
result = requests.post(list_url, json = {"submissionId":  int(sub)})
team_json = result.json()
team_df = pd.DataFrame(team_json['result']['episodes'])
print('{} games for {}'.format(len(team_df), sub))

for i in range(len(team_df)):
    epid = team_df.id.iloc[i]
    
    saveEpisode(sub, epid, team_json); r+=1;
    try:
        size = os.path.getsize('{}_{}.json'.format(sub, epid)) / 1e6
        print('Saved Episode #{} @ {:.1f}MB'.format(epid, size))
    except:
        print('file {}_{}.json did not seem to save'.format(sub, epid))    
    if r > (datetime.datetime.now() - start_time).seconds:
        time.sleep( r - (datetime.datetime.now() - start_time).seconds)
```

## Create train set

```python
import math


def log_training(result, n_machines, teams):
    """Records training data from each machine, each agent, each round
    
    Generates a training dataset to support prediction of the current
    payout ratio for a given machine.
    
    Args:
       result ([[dict]]) - output from all rounds provided as output of 
                           env.run([agent1, agent2])
       n_machines (int) - number of machines
                           
    Returns:
       training_data (pd.DataFrame) - training data, including:
           "round_num"      : round number
           "machine_id"     : machine data applies to
           "agent_id"       : player data applies to (0 or 1)
           "n_pulls_self"   : number of pulls on this machine so far by agent_id
           "n_success_self" : number of rewards from this machine by agent_id
           "n_pulls_opp"    : number of pulls on this machine by the other player
           "payout"         : actual payout ratio for this machine
    
    """
    # Initialize machine and agent states
    machine_state = [{'n_pulls_0': 0, 'n_success_0': 0, 'not_pull_since_last_0': 0, 'not_pull_since_last_1': 0,
                      'n_pulls_1': 0, 'n_success_1': 0, 'my_continue': 0, 'op_continue': 0, 'self_ucb': 0, 
                      'op_ucb': 0,
                      'payout': None}
                     for ii in range(n_machines)]
    agent_state = {'reward_0': 0, 'reward_1': 0, 'last_reward_0': 0,
                   'last_reward_1': 0}
    
    
    # Initialize training dataframe
    # - In the first round, store records for all n_machines
    # - In subsequent rounds, just store the two machines that updated
    training_data = pd.DataFrame(
            index=range(n_machines + 4 * (len(result) - 1)),
            columns=['round_num', 'machine_id', 'agent_id', 'team',
                     'n_pulls_self', 'n_success_self', 
                     'n_pulls_opp', 'n_success_opp', 'my_continue', 'op_continue', 
                     'my_not_pull', 'op_not_pull', 'explored_self', 'explored_opp', 'total_reward', 
                     'self_UCB', 'oppo_UCB', 'payout'])
    my_action_list = []
    op_action_list = []
    my_explore_set = set()
    op_explore_set = set()
    my_last_winning_list = []
    op_last_winning_list = []
    # Log training data from each round
    for round_num, res in enumerate(result):
        # Get current threshold values
        thresholds = res[0]['observation']['thresholds']

        # Update agent state
        for agent_ii in range(2):
            if res[agent_ii]['reward']:
                agent_state['last_reward_%i' % agent_ii] = (
                    res[agent_ii]['reward']
                    - agent_state['reward_%i' % agent_ii])
                agent_state['reward_%i' % agent_ii] = res[agent_ii]['reward'] 
            else: 
                agent_state['last_reward_%i' % agent_ii] = 0
                agent_state['reward_%i' % agent_ii] = 0
        # Update most recent machine state
        if res[0]['observation']['lastActions']:
            observation = res[0]['observation']
            my_idx = observation['agentIndex']
            my_last_action = observation['lastActions'][my_idx]
            op_last_action = observation['lastActions'][1-my_idx]
            
            if my_last_action and op_last_action:
                my_action_list.append(my_last_action)
                op_action_list.append(op_last_action)
                my_explore_set.add(my_last_action)
                op_explore_set.add(op_last_action)
                
            for agent_ii, r_obs in enumerate(res):
                action = r_obs['action']
                if action:
                    machine_state[action]['n_pulls_%i' % agent_ii] += 1
#                     print(machine_state[action]['n_pulls_%i' % agent_ii])
                    machine_state[action]['n_success_%i' % agent_ii] += \
                        agent_state['last_reward_%i' % agent_ii]
                    if agent_state['last_reward_%i' % agent_ii] > 0:
                        my_last_winning_list.append(action)
                    machine_state[action]['payout'] = thresholds[action]
                    for ii in range(n_machines):
                        if ii != action:
                            machine_state[ii]['not_pull_since_last_%i' % agent_ii] += 1
                        else:
                            machine_state[ii]['not_pull_since_last_%i' % agent_ii] = 0
            if len(my_action_list) >= 3:
                if my_last_action:
                    if my_action_list[-1] == my_action_list[-2]:
                        machine_state[my_last_action]['my_continue'] += 1
                    else:
                        machine_state[my_last_action]['my_continue'] = 0
            if len(op_action_list) >= 3:
                if op_last_action:
                    if op_action_list[-1] == op_action_list[-2]:
                        machine_state[op_last_action]['op_continue'] += 1
                    else:
                        machine_state[op_last_action]['op_continue'] = 0
        else:
            # Initialize machine states
            for mach_ii in range(n_machines):
                machine_state[mach_ii]['payout'] = thresholds[mach_ii]
        # Record training records
        # -- Each record includes:
        #       round_num, n_pulls_self, n_success_self, n_pulls_opp
        if res[0]['observation']['lastActions']:
            # Add results for most recent moves
            for agent_ii, r_obs in enumerate(res):
                action = r_obs['action']
                if action:
                    # Add row for agent who acted
                    row_ii = n_machines + 4 * (round_num - 1) + 2 * agent_ii 
                    training_data.at[row_ii, 'round_num'] = round_num
                    training_data.at[row_ii, 'machine_id'] = action
                    training_data.at[row_ii, 'agent_id'] = agent_ii
                    training_data.at[row_ii, 'team'] = teams[agent_ii]
                    training_data.at[row_ii, 'n_pulls_self'] = (
                        machine_state[action]['n_pulls_%i' % agent_ii])
                    training_data.at[row_ii, 'n_success_self'] = (
                        machine_state[action]['n_success_%i' % agent_ii])
                    training_data.at[row_ii, 'n_pulls_opp'] = (
                        machine_state[action]['n_pulls_%i' % (
                            (agent_ii + 1) % 2)])
                    training_data.at[row_ii, 'n_success_opp'] = (
                        machine_state[action]['n_success_%i' % (
                            (agent_ii + 1) % 2)])
                    
                    training_data.at[row_ii, 'my_continue'] = machine_state[action]['my_continue']
                    training_data.at[row_ii, 'op_continue'] = machine_state[action]['op_continue']
                    training_data.at[row_ii, 'my_not_pull'] = machine_state[action]['not_pull_since_last_%i' % agent_ii]
                    training_data.at[row_ii, 'op_not_pull'] = machine_state[action]['not_pull_since_last_%i' % (
                            (agent_ii + 1) % 2)]
                    training_data.at[row_ii, 'explored_self'] = len(my_explore_set)
                    training_data.at[row_ii, 'explored_opp'] = len(op_explore_set)
                    training_data.at[row_ii, 'total_reward'] = agent_state['reward_%i' % agent_ii]

                    average_reward = machine_state[action]['n_success_%i' % agent_ii] / (machine_state[action]['n_pulls_%i' % agent_ii] + 1e-10)
                    delta_i = math.sqrt(2 * math.log(round_num+1) / (machine_state[action]['n_pulls_%i' % agent_ii] + 1e-10))
                    upper_bound = average_reward + delta_i

                    training_data.at[row_ii, 'self_UCB'] = upper_bound
                    
                    average_reward = machine_state[action]['n_success_%i' % ((agent_ii + 1) % 2)] / (machine_state[action]['n_pulls_%i' % ((agent_ii + 1) % 2)] + 1e-10)
                    delta_i = math.sqrt(2 * math.log(round_num+1) / (machine_state[action]['n_pulls_%i' % ((agent_ii + 1) % 2)] + 1e-10))
                    upper_bound = average_reward + delta_i
                    
                    training_data.at[row_ii, 'oppo_UCB'] = upper_bound
                    
                    training_data.at[row_ii, 'payout'] = (
                        machine_state[action]['payout'] / 100)

                    # Add row for other agent
                    row_ii = n_machines + 4 * (round_num - 1) + 2 * agent_ii + 1
                    other_agent = (agent_ii + 1) % 2
                    training_data.at[row_ii, 'round_num'] = round_num
                    training_data.at[row_ii, 'machine_id'] = action
                    training_data.at[row_ii, 'agent_id'] = other_agent
                    training_data.at[row_ii, 'team'] = teams[other_agent]
                    training_data.at[row_ii, 'n_pulls_self'] = (
                        machine_state[action]['n_pulls_%i' % other_agent])
                    training_data.at[row_ii, 'n_success_self'] = (
                        machine_state[action]['n_success_%i' % other_agent])
                    training_data.at[row_ii, 'n_pulls_opp'] = (
                        machine_state[action]['n_pulls_%i' % agent_ii])
                    training_data.at[row_ii, 'n_success_opp'] = (
                        machine_state[action]['n_success_%i' % agent_ii])
                    training_data.at[row_ii, 'my_continue'] = machine_state[action]['my_continue']
                    training_data.at[row_ii, 'op_continue'] = machine_state[action]['op_continue']
                    training_data.at[row_ii, 'my_not_pull'] = machine_state[action]['not_pull_since_last_%i' % other_agent]
                    training_data.at[row_ii, 'op_not_pull'] = machine_state[action]['not_pull_since_last_%i' % agent_ii]
                    training_data.at[row_ii, 'explored_self'] = len(op_explore_set)
                    training_data.at[row_ii, 'explored_opp'] = len(my_explore_set)
                    training_data.at[row_ii, 'total_reward'] = agent_state['reward_%i' % other_agent]
                    
                    average_reward = machine_state[action]['n_success_%i' % other_agent] / (machine_state[action]['n_pulls_%i' % other_agent] + 1e-10)
                    delta_i = math.sqrt(2 * math.log(round_num+1) / (machine_state[action]['n_pulls_%i' % other_agent] + 1e-10))
                    upper_bound = average_reward + delta_i

                    training_data.at[row_ii, 'self_UCB'] = upper_bound
                    
                    average_reward = machine_state[action]['n_success_%i' % agent_ii] / (machine_state[action]['n_pulls_%i' % agent_ii] + 1e-10)
                    delta_i = math.sqrt(2 * math.log(round_num+1) / (machine_state[action]['n_pulls_%i' % agent_ii] + 1e-10))
                    upper_bound = average_reward + delta_i
                    
                    training_data.at[row_ii, 'oppo_UCB'] = upper_bound
                    
                    training_data.at[row_ii, 'payout'] = (
                        machine_state[action]['payout'] / 100)
                
        else:
            # Add initial data for all machines
            for action in range(n_machines):
                row_ii = action
                training_data.at[row_ii, 'round_num'] = round_num
                training_data.at[row_ii, 'machine_id'] = action
                training_data.at[row_ii, 'agent_id'] = -1
                training_data.at[row_ii, 'team'] = -1
                training_data.at[row_ii, 'n_pulls_self'] = 0
                training_data.at[row_ii, 'n_success_self'] = 0
                training_data.at[row_ii, 'n_pulls_opp'] = 0
                training_data.at[row_ii, 'n_success_opp'] = 0
                training_data.at[row_ii, 'my_continue'] = 0
                training_data.at[row_ii, 'op_continue'] = 0
                training_data.at[row_ii, 'my_not_pull'] = 0
                training_data.at[row_ii, 'op_not_pull'] = 0
                training_data.at[row_ii, 'explored_self'] = 0
                training_data.at[row_ii, 'explored_opp'] = 0
                training_data.at[row_ii, 'total_reward'] = 0
                training_data.at[row_ii, 'self_UCB'] = 0
                training_data.at[row_ii, 'oppo_UCB'] = 0
                training_data.at[row_ii, 'payout'] = (
                    machine_state[action]['payout'] / 100)
            
    return training_data
```
